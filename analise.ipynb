```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura dos dados do arquivo
df = pd.read_csv("lin3/zero_lin_3/zero_lin_3.txt", sep=r"\s+\|\s+|\s+", engine="python", header=None)

title= "População inicial de 0's / Função de resfriamento linear"

# Renomeando as colunas
df.columns = ["valor", "peso", "iteracao", "ultimoValor", "ultimoPeso"]

# Exibindo o DataFrame
print(df)
```

```python
# Encontrar o maior valor na coluna 'valor'
maior_valor = df["valor"].max()

# Contar quantas vezes o maior valor aparece
ocorrencias_maior_valor = (df["valor"] == maior_valor).sum()

# Calcular a porcentagem de vezes que o maior valor aparece
percentual_maior_valor = (ocorrencias_maior_valor / len(df)) * 100

# Exibindo o maior valor
print("O maior valor é:", maior_valor)
```

```python
# Normalizando os valores da coluna 'valor' para uma escala de 0 a 100%
df["valor_percentual"] = (df["valor"] / df["valor"].max()) * 100

# Contando as ocorrências abaixo de 80% do valor máximo
ocorrencias_abaixo_80 = (df["valor_percentual"] < 80).sum()
percentual_abaixo_80 = (ocorrencias_abaixo_80 / len(df)) * 100
```

```python
# Normalizando os valores da coluna 'valor' para uma escala de 0 a 100%
df["valor_percentual"] = (df["valor"] / df["valor"].max()) * 100

# Criando o histograma
plt.hist(df["valor_percentual"], bins=40, edgecolor="black")  # 10 bins para divisão em faixas de valores
plt.xlabel("Valor (% do máximo)")
plt.ylabel("Frequência")
plt.title(title)

# Ajustando o eixo x para mostrar todo o intervalo de 0 a 100%
plt.xlim(0, 100)

# Adicionando a legenda com a porcentagem de vezes que o maior valor aparece
plt.legend([f"Maior valor aparece {percentual_maior_valor:.2f}% das vezes\n\nValores abaixo de 80%: {percentual_abaixo_80:.2f}%"])

# Exibindo o gráfico
plt.show()
```

```python
# Normalizando os valores da coluna 'valor' para uma escala de 0 a 100%
df["valor_percentual_ultimo"] = (df["ultimoValor"] / df["valor"].max()) * 100

# Criando o histograma
plt.hist(df["valor_percentual_ultimo"], bins=10, edgecolor="black")  # 10 bins para divisão em faixas de valores
plt.xlabel("último valor (% do máximo)")
plt.ylabel("Frequência")
plt.title(title)

# Ajustando o eixo x para mostrar todo o intervalo de 0 a 100%
plt.xlim(0, 100)

# Exibindo o gráfico
plt.show()
```

```python
# Calculando o número médio de iterações
media_iteracoes = df["iteracao"].mean()

plt.figure(figsize=(10, 6))
plt.scatter(df.index, df["iteracao"], color="blue", alpha=0.6, label="Iterações")
plt.axhline(y=media_iteracoes, color="red", linestyle="--", label=f"Média de Iterações: {media_iteracoes:.2f}")
plt.title("Iterações Necessárias por execução")
plt.xlabel("Execuções")
plt.ylabel("Número de Iterações")

# Colocando a legenda na parte inferior
plt.legend(loc="lower right", bbox_to_anchor=(1, -0.15), fancybox=True, shadow=True, ncol=2)
plt.grid(True)
plt.tight_layout()
plt.show()
```
